```python
x = dataset.iloc[:, [3, 4]].values
```

```python
# Finding optimal number of clusters using the elbow method
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss_list = []  # Initializing the list for the values of WCSS

# Using a for loop for iterations from 1 to 10.
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss_list)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

```python
# Training the K-means model on a dataset
from sklearn.cluster import KMeans

# Assuming 'x' is already defined as the dataset features
# kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
# y_predict = kmeans.fit_predict(x)

# The code seems correct, but let's ensure 'x' is defined and contains the correct data
# For demonstration, let's assume 'x' is a 2D array with numerical data
# If 'x' is not defined, you would need to define it here, for example:
# x = some_data_source

# Now, let's train the K-means model with the assumed 'x'
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_predict = kmeans.fit_predict(x)

# Print the cluster centers and the predicted clusters for the dataset
print("Cluster centers:\n", kmeans.cluster_centers_)
print("Predicted clusters for the dataset:\n", y_predict)

# Note: Replace 'some_data_source' with the actual data source or dataset variable
```

```python
import matplotlib.pyplot as plt

# Assuming 'x' and 'y_predict' are already defined
# If not, replace them with your actual data

# Scatter plots for each cluster
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s=100, c='black', label=' Average Salary and Spending Customer  ')  # First cluster
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s=100, c='orange', label='Careful Customer')  # Second cluster
plt.scatter(x[y_predict == 2, 0], x[y_predict == 2, 1], s=100, c='pink', label=' Sensible Customer')  # Third cluster
plt.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], s=100, c='red', label='Careless Customer')  # Fourth cluster
plt.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s=100, c='green', label='Target Customer Customer')  # Fifth cluster

# Scatter plot for centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='yellow', label='Centroid')

# Plot labels and title
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

# Show legend
plt.legend()

# Display the plot
plt.show()
```
