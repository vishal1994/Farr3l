```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create a sample DataFrame with weight and height data
data = pd.DataFrame({
  "Weight": [185, 170, 168, 179, 182, 188, 180, 180, 183, 180, 177, 177],
  "Height": [72, 56, 60, 68, 72, 77, 71, 70, 84, 88, 67, 76]
})

# Exploratory Data Analysis (EDA)

# Descriptive statistics for weight and height
print(data.describe())

# Scatter plot of weight vs height
plt.figure(figsize=(8, 6))
plt.scatter(data["Weight"], data["Height"], color='blue')
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Weight vs Height Distribution")
plt.grid(True)
plt.show()

# K-Means Clustering

# Define the number of clusters (k)
k = 2  # You can adjust this value

# Create the KMeans model
kmeans = KMeans(n_clusters=k, random_state=0)

# Fit the model to the data
kmeans.fit(data)

# Get cluster labels for each data point
y_kmeans = kmeans.labels_

# Scatter plot with points colored by cluster assignments
plt.figure(figsize=(8, 6))
plt.scatter(data["Weight"], data["Height"], c=y_kmeans, cmap='viridis')
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Weight vs Height with K-Means Clusters (k=" + str(k) + ")")
plt.grid(True)
plt.show()

# Print cluster centroids
print("Cluster centroids:\n", kmeans.cluster_centers_)
```
